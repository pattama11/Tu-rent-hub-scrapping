```python
import requests as req
import pandas as pd
from bs4 import BeautifulSoup
import warnings
import numpy as np
```

```python
web = 'https://www.renthub.in.th/%E0%B8%AD%E0%B8%9E%E0%B8%B2%E0%B8%A3%E0%B9%8C%E0%B8%97%E0%B9%80%E0%B8%A1%E0%B9%89%E0%B8%99%E0%B8%97%E0%B9%8C-%E0%B8%AB%E0%B9%89%E0%B8%AD%E0%B8%87%E0%B8%9E%E0%B8%B1%E0%B8%81-%E0%B8%AB%E0%B8%AD%E0%B8%9E%E0%B8%B1%E0%B8%81/%E0%B8%A1-%E0%B8%98%E0%B8%A3%E0%B8%A3%E0%B8%A1%E0%B8%A8%E0%B8%B2%E0%B8%AA%E0%B8%95%E0%B8%A3%E0%B9%8C-%E0%B8%A8%E0%B8%B9%E0%B8%99%E0%B8%A2%E0%B9%8C%E0%B8%A3%E0%B8%B1%E0%B8%87%E0%B8%AA%E0%B8%B4%E0%B8%95'
```

```python
import urllib.parse
url = [web]
res = req.get(web).content
soup = BeautifulSoup(res, 'html.parser')
ul2 = soup.find('ul',class_='css-jl2bni')
for i in ul2 :
    for link in i.find_all('a', href=True):
        link['href'] = '' + link['href'][1:]
        link['href'] = 'https://www.renthub.in.th/' + urllib.parse.quote(link['href'])
        if link['href'] not in url:
            url.append(link['href'])
```

```python
title = []
p_m=[]
p_d=[]
dist = []
addr = []
link = []
for u in url:
    res = req.get(u).content
    soup = BeautifulSoup(res, 'html.parser')
    ul = soup.find('ul',class_='css-1i8gfqr')
    li = ul.find_all('li')
    for e in li:
        title.append(e.find('span',class_='css-7wvhf9').text)
        l = e.find('a', class_='css-10i63lj').get('href')
        link.append('https://www.renthub.in.th/' + urllib.parse.quote(l))
        addr.append(e.find('span',class_='address css-8od65o').text)
        if e.find('span',class_='css-1n32v8m') != None:
            dist.append(e.find('span',class_='css-1n32v8m').text)
        else :
            dist.append(None)
        price = e.find('div', class_='css-1sky5v4').find_all('div')
        if len(price)>1:
            for i in range(len(price)):
                if 'เดือน' in price[i].text:
                    p_m.append(price[i].text)
                elif 'วัน' in price[i].text:
                    p_d.append(price[i].text)
        else :
            if 'เดือน' in price[0].text:
                p_m.append(price[0].text)
                p_d.append(None)
            elif 'วัน' in price[0].text:
                p_d.append(price[0].text)
                p_m.append(None)
df = pd.DataFrame({'name':title,'price/day' : p_d, 'price/month' : p_m,'distance' : dist,'address' : addr, 'link' : link})
```

```python
warnings.simplefilter('ignore')
df['name'] = df['name'].str.replace('อพาร์ทเม้นท์ หอพัก ย่าน ม.ธรรมศาสตร์ ศูนย์รังสิต : ','')
```

```python
df.info()
```

```python
df['price/day'] = df['price/day'].str.replace('บาท/วัน','')
df['price/day'] = df['price/day'].str.replace('รายวัน :','')
df['lowest_price/day'] = df['highest_price/day'] = df['price/day']
for i in range(len(df['price/day'])):
    if df['price/day'][i] != None:
        if ' - ' in df['price/day'][i]:
            df['lowest_price/day'][i] = df['price/day'][i].split(' - ')[0]
            df['highest_price/day'][i] = df['price/day'][i].split(' - ')[1]
        elif 'โทรสอบถาม' in df['price/day'][i]:
            df['lowest_price/day'][i] = df['highest_price/day'][i] = np.nan
    else:
        df['lowest_price/day'][i] = df['highest_price/day'][i] = np.nan
df = df.drop(columns='price/day')
```

```python
df['price/month'] = df['price/month'].str.replace('บาท/เดือน','')
df['price/month'] = df['price/month'].str.replace('รายเดือน :','')
df['lowest_price/month'] = df['highest_price/month'] = df['price/month']
for i in range(len(df['price/month'])):
    if df['price/month'][i] != None:
        if ' - ' in df['price/month'][i]:
            df['lowest_price/month'][i] = df['price/month'][i].split(' - ')[0]
            df['highest_price/month'][i] = df['price/month'][i].split(' - ')[1]
        elif 'โทรสอบถาม' in df['price/month'][i]:
            df['lowest_price/month'][i] = df['highest_price/month'][i] = np.nan
    else:
        df['lowest_price/month'][i] = df['highest_price/month'][i] = np.nan
df = df.drop(columns='price/month')
```

```python
df['distance'] = df['distance'].str.replace('ห่างประมาณ ','')
df['distance'] = df['distance'].str.replace('ห่างออกไป ','')
for i in range(len(df['distance'])):
    if df['distance'][i] != None :
        if ' เมตร' in df['distance'][i] :
            df['distance'][i] = df['distance'][i].replace(' เมตร','')
            df['distance'][i] = float(df['distance'][i])/1000
        else:
            df['distance'][i] = df['distance'][i].replace(' กม.','')
    else:
        df['distance'][i] = np.nan
```

```python
cols = ['name', 'lowest_price/day', 'highest_price/day', 'lowest_price/month','highest_price/month','distance','address','link']
df = df[cols]
df
```
